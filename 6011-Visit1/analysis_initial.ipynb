# Initial Statistical Testing on TEWL (as a reference, skip to permutation test for more robust testing)

## paired t-test (assuming normal distribution)


- compare the means of the same group or item under two separate scenarios
- assumption might not hold true - see wilcoxon signed ranked test for more
- test whether there is significant difference between points related to treatment A compared to B

```python
ttest_results = []
ttest_pvalues = []
for i in range(num_par): ## loop through each subject
    test_df = df3[df3['id']== i+1] ## test_df contains data for treatment A and B for a single subject
    result, pvalue = ttest_rel(test_df.iloc[0,2:],test_df.iloc[1,2:]) ## test difference between 4 data points by A and 4 data points by B
    ttest_results += [result]
    ttest_pvalues += [pvalue]
## the operation above generates a ttest result and pvalue for each subject
## the code below describes the pvalues distribution
describe(ttest_pvalues)  
```

Data doesn't really satisfy the assumption of this method

## wilcoxon signed ranked test (alternative to paired t test)

test whether there is significant difference between points related to treatment A compared to B

```python
wsr_results = []
wsr_pvalues = []
significant_ind = []
high_cor =[]
middle_cor = []
## repeat the same operation as paired t-test
for i in range(num_par):
    test_df = df3[df3['id']== i+1]
    result, pvalue = wilcoxon(test_df.iloc[0,2:],test_df.iloc[1,2:])
    if pvalue <= 0.20: ## 0.2 cutoff too tight
        significant_ind += [i+1]
    ## if correlation is >= 0.6, categorize as high correlation between data for A and B
    elif pvalue >= 0.6: 
        high_cor += [i+1]
    else:
        middle_cor += [i+1]
    wsr_results  += [result]
    wsr_pvalues += [pvalue]
significant_ind
```

```python
from scipy.cluster.hierarchy import dendrogram
def plot_dendrogram(model, **kwargs):
    # Create linkage matrix and then plot the dendrogram

    # create the counts of samples under each node
    counts = np.zeros(model.children_.shape[0])
    n_samples = len(model.labels_)
    for i, merge in enumerate(model.children_):
        current_count = 0
        for child_idx in merge:
            if child_idx < n_samples:
                current_count += 1  # leaf node
            else:
                current_count += counts[child_idx - n_samples]
        counts[i] = current_count

    linkage_matrix = np.column_stack([model.children_, model.distances_,
                                      counts]).astype(float)

    # Plot the corresponding dendrogram
    dendrogram(linkage_matrix, **kwargs)
```

```python
model = AgglomerativeClustering(linkage='complete',distance_threshold=0, n_clusters=None)
model = model.fit(dfA.T)
plot_dendrogram(model)
plt.title('hierarchical clustering for test article A')
```
